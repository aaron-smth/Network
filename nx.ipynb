```python
import networkx as nx
import matplotlib.pyplot as plt
import collections
from scipy.stats import linregress
import numpy as np
```

```python
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (20, 10),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)
plt.style.use('fivethirtyeight')
```

```python
def load_data(fname):
    info = {}
    with open(fname, 'rb') as f:
        G = nx.read_edgelist(f)
    info['G'] = G
    degree_seq = sorted([d for n, d in G.degree()], reverse=True)
    degree_arr = np.array(degree_seq)
    info['degree_mean'] = degree_arr.mean()
    info['degree_std'] = degree_arr.std()
    info['degree_count'] = collections.Counter(degree_seq)
    info['degree_correlation'] = nx.degree_pearson_correlation_coefficient(G)
    return info
```

```python
def degree_dist(degree_count, fname, **kwargs):
    fig, ax = plt.subplots()
    xdata, ydata = np.fromiter(degree_count.keys(), dtype=float), np.fromiter(degree_count.values(), dtype=float)
    ax.scatter(xdata, ydata, label='raw data')
    trunc = (xdata<200) & (xdata>10)
    xdata_truncate, ydata_truncate = xdata[trunc], ydata[trunc]
    reg = linregress(np.log(xdata_truncate), np.log(ydata_truncate))
    
    fitted_ydata = np.exp(np.log(xdata) * reg.slope + reg.intercept)
    ax.plot(xdata, fitted_ydata, label='fitted line', color='red')
    ax.annotate(f'fitted slope: {reg.slope:.{3}f}', xy=(0.7,0.7), xycoords='axes fraction', fontsize=20)
    ax.annotate(f'mean of degree: {kwargs["mean"]:.{3}f}', xy=(0.7,0.6), xycoords='axes fraction', fontsize=20)
#     ax.annotate(f'std of degree: {kwargs["std"]:.{3}f}', xy=(0.7,0.5), xycoords='axes fraction', fontsize=20)
    
    ax.set(xscale='log', yscale='log', title=f'{fname[:-4]} Network Degree distribution', 
           xlabel='DEGREE k', ylabel='NODES WITH DEGREE k')
    
    ax.legend(fontsize=20)
    plt.xticks(size=20)
    plt.yticks(size=20)
    plt.show(fig)
    fig.savefig(f'deg_dist_{fname[:-4]}.png')
```

```python
fname_li = ['facebook_combined.txt', 'p2p-Gnutella04.txt', 'email-Enron.txt']
```

```python
for fname in fname_li:
    info = load_data(fname)
    degree_dist(info['degree_count'], fname, mean=info['degree_mean'], std=info['degree_std'])
```

```python
def small_world(G, **kwargs):
    name = kwargs['fname'][:-4]
    mean_dist_li = []
    n_samples = 500
    for node in np.random.choice(G.node, size=n_samples):
        dist = nx.single_source_shortest_path_length(G, node).values()
        if len(dist) == 1:
            continue
        mean_dist_li.append(sum(dist) / len(dist))
    fig, ax = plt.subplots()
    ax.hist(mean_dist_li, bins='sqrt')
    ax.set(xlabel='average shortest path length to all other nodes', ylabel='frequency', title=f'{name} average path length distribution')
    mean_dist_arr = np.array(mean_dist_li)
    avg, std, max_, min_ = mean_dist_arr.mean(), mean_dist_arr.std(), mean_dist_arr.max(), mean_dist_arr.min()
    ax.annotate(f'number of samples:{n_samples}', xy=(0.75,0.9), xycoords='axes fraction', fontsize=20)
    ax.annotate(f'mean: {avg:.{3}f}, std: {std:.{3}f}', xy=(0.75,0.8), xycoords='axes fraction', fontsize=20)
    ax.annotate(f'max: {max_:.{3}f}, min: {min_:.{3}f}', xy=(0.75,0.7), xycoords='axes fraction', fontsize=20)
    plt.show(fig)
    fig.savefig(f'path_length_{name}.png')
```

```python
for fname in fname_li:
    info = load_data(fname)
    small_world(info['G'], fname=fname)
```

```python
def connected_node_ratio(graph, centers):
    centers = [center for center in centers if graph.has_node(center)]
    nodes = set()
    for center in centers:
        nodes.update(nx.single_source_shortest_path(graph, center, cutoff=4).keys())
    return len(nodes) / graph.number_of_nodes()

def resilience_test(G, random=True):
    graph = G.copy()
    deg_centrality = nx.degree_centrality(graph)
    print('calculating centers')
    centers_tup_all = sorted(deg_centrality.items(), key=lambda tup:tup[1], reverse=True)
    centers_all = [tup[0] for tup in centers_tup_all]
    centers = centers_all[:10]
    
    d_removal_percent = 0.02 * 0.01
    d_removal_n = int(graph.number_of_nodes() * d_removal_percent)
    n_samples = 100
    
    removal_ratio_li = [0.]
    connection_ratio_li = [connected_node_ratio(graph, centers)]
    # 25 * d_removal_percent = total_removal_percent
    print('calculating nodes')
    for i in range(25):
        if random == True:
            choices = graph.nodes()
        else:
            choices = centers_all[:100]
        graph.remove_nodes_from(np.random.choice(choices, size=d_removal_n))
        rate = (i+1) * d_removal_percent
        ratio = connected_node_ratio(graph, centers)
        removal_ratio_li.append(rate)
        connection_ratio_li.append(ratio)
        print(f'removal ratio:{rate}, connection ratio:{ratio}')
    return removal_ratio_li, connection_ratio_li

def plot_resilience(removal_ratio_li, connection_ratio_li, **kwargs):
    fname = kwargs['fname'][:-4]
    fig, ax = plt.subplots()
    ax.scatter(removal_ratio_li, connection_ratio_li, label='raw data')
    ax.set(ylabel='perentage of nodes connected to a few centers',
           xlabel='percentage of nodes removed from the network', title=f'{fname} resilience test (targeting centers)')
    
    xdata, ydata = np.array(removal_ratio_li), np.array(connection_ratio_li)
    reg = linregress(xdata, ydata)
    
    fitted_ydata = xdata * reg.slope + reg.intercept
    ax.plot(xdata, fitted_ydata, label='fitted line', color='red')
    ax.annotate(f'fitted slope: {reg.slope:.{3}f}', xy=(0.7,0.7), xycoords='axes fraction', fontsize=20)
                  
    ax.set(xlim=(xdata.min(), xdata.max()+0.001))
    ax.legend()
    plt.show(fig)
    fig.savefig(f'resilience2_{fname}.png')
```

```python
for fname in fname_li:
    info = load_data(fname)
    removal_ratio_li, connection_ratio_li = resilience_test(G, random=False)
    plot_resilience(removal_ratio_li, connection_ratio_li, fname=fname)
```
